```python
data_title = [] #taruh di awal supaya gak ketumpuk
data_classification = []
data_address = []
data_services = []
for index in range (1,11): #pakai looping 
    page_index = requests.get(f'https://www.yell.com/ucs/UcsSearchAction.do?keywords=Hairdressers&location=UK&scrambleSeed=938654611&pageNum={index}') #index akan berjalan sesuai halaman#f artinya supaya mau berubah ubah
    soup_index = BeautifulSoup(page_index.content, 'html.parser') #buat merapikan
    
    title_html = soup.find_all('div', class_='businessCapsule--titSpons') #ambil semua title yang memuat di halaman ke-n
    for each_title_html in title_html:
        data_title.append(each_title_html.text.strip())
        
    classification_html = soup.find_all('span', class_='businessCapsule--classification')
    for each_classification_html in classification_html:
        data_classification.append(each_classification_html.text.strip())
        
    address_html = soup.find_all('a', class_='col-sm-24 businessCapsule--address businessCapsule--link') #ambil semua address yang memuat di halaman ke-n
    for each_address_html in address_html:
        data_address.append(each_address_html.text.strip())
        
    services_html = soup.find_all('div', class_='businessCapsule--services') #ambil semua services yang memuat di halaman ke-n
    for each_services_html in  services_html:
        data_services.append(each_services_html.text.strip())
```
